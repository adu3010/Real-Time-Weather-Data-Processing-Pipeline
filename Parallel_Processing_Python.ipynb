#Simulate Random Data

```python
import time
import random
import pandas as pd

# Generate sample data
def generate_weather_data():
    cities = ['CityA', 'CityB', 'CityC']
    while True:
        data = {
            'City': random.choice(cities),
            'Temperature': random.uniform(15, 35),
            'Humidity': random.uniform(30, 90),
            'Timestamp': pd.Timestamp.now()
        }
        yield data
        time.sleep(1)  # Simulates 1-second data intervals
```

#Implement Multi-Threading for Real-Time Processing

```python
import threading
import queue
import matplotlib.pyplot as plt

# Shared queue for processing
data_queue = queue.Queue()

# Producer thread: Simulates data stream
def producer():
    for data in generate_weather_data():
        print(f"Produced: {data}")
        data_queue.put(data)

# Consumer thread: Processes the data
def consumer():
    processed_data = []
    while True:
        if not data_queue.empty():
            data = data_queue.get()
            # Data cleaning/aggregation
            data['Temperature_Celsius'] = round((data['Temperature'] - 32) * 5/9, 2)
            processed_data.append(data)
            print(f"Consumed: {data}")
            # Optional: Visualize real-time updates
            if len(processed_data) % 10 == 0:
                visualize(processed_data)

# Visualization function
def visualize(data):
    df = pd.DataFrame(data)
    plt.plot(df['Timestamp'], df['Temperature_Celsius'], label='Temperature (Celsius)')
    plt.xlabel('Time')
    plt.ylabel('Temperature')
    plt.legend()
    plt.show()

# Start threads
producer_thread = threading.Thread(target=producer, daemon=True)
consumer_thread = threading.Thread(target=consumer, daemon=True)

producer_thread.start()
consumer_thread.start()

# Keep the main thread alive
while True:
    time.sleep(1)
```

#Optimise with Multi-Processing
For CPU-bound tasks like heavy computations, we replace threading with multiprocessing.

Adjustments for Multi-Processing:

Replace queue.Queue with multiprocessing.Queue.

Use the multiprocessing.Process class instead of threading.Thread

```python
import multiprocessing

# Use multiprocessing.Queue
data_queue = multiprocessing.Queue()

# Adjust producer and consumer for multiprocessing
producer_process = multiprocessing.Process(target=producer)
consumer_process = multiprocessing.Process(target=consumer)

# Start processes
producer_process.start()
consumer_process.start()

# Wait for processes to complete (optional for non-daemon)
producer_process.join()
consumer_process.join()
```
